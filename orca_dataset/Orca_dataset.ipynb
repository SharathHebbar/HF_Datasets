```python
def clean_dataset(example):
    system = example['system_prompt']
    user = example['question']
    response = example['response']
    if system is None or system == "":
        system = "You are an AI assistant and expertise in everything"
    example['text'] = f"""<|system|>\n{system}</s>.\n<|user|>\n{user}\n<|assistant|>\n{response}"""
    return example
```

```python
dataset = dataset.map(
    clean_dataset, remove_columns=['id', 'system_prompt', 'question', 'response']
)
```
