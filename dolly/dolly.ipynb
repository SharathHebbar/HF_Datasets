```python
from datasets import load_dataset

dataset = load_dataset("databricks/databricks-dolly-15k", split='train')
```

```python
dataset
```

```python
def clean_dataset(data):
  system = data['context']
  user = data['instruction']
  response = data['response']
  data['prompt'] = f"""<|system|>\n{system}</s>.\n<|user|>\n{user}\n<|assistant|>\n{response}"""
  return data

dataset = dataset.map(
    clean_dataset, remove_columns=['instruction', 'context', 'category', 'response']
)
```
