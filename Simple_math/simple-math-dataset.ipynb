```python
from datasets import load_dataset, DatasetDict, concatenate_datasets

train_dataset = load_dataset("fblgit/simple-math", split='train')
test_dataset = load_dataset("fblgit/simple-math", split='test')
```

```python
test_dataset
```

```python
train_dataset['instruction'][0]
```

```python
def clean_dataset(example):
    example['prompt'] = f"The output of {example['instruction']} is {example['output']}"
    return example

train_dataset = train_dataset.map(
    clean_dataset, remove_columns=['instruction', 'output']
)

test_dataset = test_dataset.map(
    clean_dataset, remove_columns=['instruction', 'output']
)
```

```python
dataset = concatenate_datasets([train_dataset, test_dataset])
dataset
```

```python
datasets = DatasetDict({
    'train': train_dataset,
    'test': test_dataset,
    'all': dataset
})
```
