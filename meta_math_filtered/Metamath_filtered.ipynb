```python
def clean_dataset(data):
  # system = data['instruction']
  system = None
  user = data['query']
  response = data['response']
  if system is None or system == "":
    system = "You are an AI Assistant expertise in math"
  if user is None or user == "":
    user = "Can you answer based on context?"
  data['prompt'] = f"""<|system|>\n{system}</s>.\n<|user|>\n{user}\n<|assistant|>\n{response}"""
  return data

dataset = dataset.map(
    clean_dataset, remove_columns=['query', 'response']
)
```
