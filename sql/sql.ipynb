```python
from datasets import load_dataset

dataset = load_dataset("b-mc2/sql-create-context", split='train')
```

```python
dataset
```

```python
def clean_dataset(data):
  system = data['context']
  user = data['question']
  response = data['answer']
  if system is None or system == "":
    system = "Can you answer this?"
  data['prompt'] = f"""<|system|>\n{system}</s>.\n<|user|>\n{user}\n<|assistant|>\n{response}"""
  return data

dataset = dataset.map(
    clean_dataset, remove_columns=['context', 'question', 'answer']
)
```
